# Transpose and Swap Arrays

```python
# Import needed libraries

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create array of numbers

array = np.arange(0,10).reshape(2,5)
array
```

```python
# Reshape array to (5,2)

array_reshaped = array.reshape(5,2)
array_reshaped
```

```python
# Add arrays

array_add = np.add(array,array_reshaped.T)
array_add
```

```python
# Swap indexes of array_add

array_swap_index = np.swapaxes(array_add,axis1=0,axis2=1)
array_swap_index
```

```python
# Slice of array_swap_index

slice_swap_index = array_swap_index[0]
slice_swap_index
```

```python
# Lists of inputs

players = ["Robin", "Leo", "Pogba", "Diego", "Ronaldo"] 
teams = ["Man UTD", "Barca", "Juve", "Napoli", "R Madrid"]
goals = [[12, 15,16,15,13],
         [26,30,31,25,24],
         [10,12,8,6,13],
         [18,19,17,20,21],
         [21,32,25,21,22]
         ]
```

```python
# Create array of goals
array = np.array(goals)
array
```

```python
# Create heatmap of goals

fig,ax = plt.subplots(figsize=(12,8))
sns.heatmap(data=array,
            annot= True
            )
ax.set_xticks(np.arange(len(players)), labels=players)
ax.set_yticks(np.arange(len(teams)), labels=teams)
plt.title("Player Analysis Heatmap", fontsize=20)
plt.show()
```
