 # NumPy Arrays and Vector Operations


```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating arrays
rng = np.random.default_rng(seed=42)

array1 = rng.integers(low=10,high=21,size=(2,3))
array2 = rng.integers(low=10,high=21,size=(1,3))

# array2
```

```python
# Checking shape of arrays

print(f'Shape of array1 is {array1.shape}') 
print(f'Shape of array2 is {array2.shape}')
```

```python
# Adding the two arrays

array3 = np.add(array1, array2)
array3
```

```python
#   Finding the dot product of the two arrays

dot_product = np.dot(array1, array2.transpose())
dot_product
```

```python
# 1-dimensional array

array = rng.integers(low=0, high=11,size=(100))
# array
```

```python
# Histo plot of 1-dimensional array

fig,ax = plt.subplots(figsize=(10,8))
ax.hist(array,bins=10)
ax.set(title="Histogram of 1-dimensional array")
plt.show()
```

```python
# Compute mean and median 

mean = np.mean(array)
median = np.median(array)

print(f'The mean and median are {mean} and {median} respectively')
```

```python
from statistics import mode

mode = mode(array)
print(f'The mode of 1-dimensional array is {mode}')
```

```python
# Create the 3d array
array_3d = rng.integers(low=0,high=11,size=(100,3))
# array_3d
```

```python
# 3D scatterplot of 3d array

fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(projection='3d')

ax.scatter(
    array_3d[:,0],
    array_3d[:,1],
    array_3d[:,2]
)
plt.title('3D Scatter Plot')
plt.show()
```
