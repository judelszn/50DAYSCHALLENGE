```python
# Import necessary libraries

import pandas as pd
import numpy as np
```

```python
names = ["Joe", "Phil", "Ken", "Jos", "Luke"]
miles_run = [120,80,100,90,85]
time_in_hrs = [40,38,45,50,50]

# Create dataframe with lists above
df = pd.DataFrame({'Name':names, 'Miles Run':miles_run, 'Time':time_in_hrs})
df.head()
```

```python
names = ["Joe", "Phil", "Ken", "Jos", "Luke"]
age = [45,28,21,55,62]
gender = ["M","M","F","F","M"]

# Create a new data frame

df2 = pd.DataFrame({"Name": names, "Age": age, "Gender": gender})
df2.head()
```

```python
# Merge the two data frame
df_merged = pd.concat([df, df2.drop(columns="Name")], axis=1)
df_merged.head()
```

```python
# Reshuffle the data frame
df = df_merged.sample(frac=1)
df.head()
```

### Why is important to reshuffle data frame for machine learning?

- To prevent overfitting with other issues

```python
# Melt the data frame into long format

df_long = df.melt(id_vars="Name",
                  var_name="Variables",
                  value_name="Values"
                  )
df_long.head()
```
