# Day 3: Generating Random Arrays

```python
# Importing needed libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Array of random numbers
rng = np.random.default_rng(seed = 42)
array = rng.random(size=(3,4))
array
```

```python
rng = np.random.default_rng()
array = rng.integers(10,size=(3,4))
array
```

```python
# Array of random numbers with standard normal distribution
array = rng.standard_normal(size=1000)
#array
```

```python
# Plot histogram of the distribution
fig,ax = plt.subplots(figsize=(10,8))
ax.hist(array,bins=50)
ax.set(title="Standard Normal distribution")
plt.show()
```

```python
arr = [ "Orange", "Apple", "Pear"]

# Array from the list above
array = rng.choice(arr,size=(3,4))
array
```
