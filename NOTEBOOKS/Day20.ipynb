#  Explore Data with Pandas and Matplotlib

```python
# Import necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
products = ["sugar", "Salt", "oil", "diapers", "rice"]
costs = [2450,1989,6745,9807,8743]
sales = [27908,4508,6734,9976,9000]

# Create a data frame with lists above
df = pd.DataFrame({'Products': products,'Costs': costs,'Sales': sales})
df.head()
```

```python
# Find most profitable product

# Create of profit
df["Profit"] = df["Sales"] - df["Costs"]

# Index of most profitable product
index_profit_product = df["Profit"].idxmax()

most_profit_product = df.loc[index_profit_product,"Products"]

print(f"The most profitable product is {most_profit_product}")
```

```python
# Bar plot of profits of products

fig,ax = plt.subplots(figsize=(12,8))
ax.bar(height=df["Profit"],
       x=df["Products"])
ax.set(xlabel="Products on Sales",ylabel="Profit Gain")
plt.title(label="Product Profit Analysis", fontdict={"fontsize":20,"fontweight":"bold"})
plt.show()
```

```python
# Find the least profitable product

#index of leasr profitable product
index_least_profit_product = df["Profit"].idxmin()

least_profit_product = df.loc[index_least_profit_product,"Products"]
print(f"The least profitable product is {least_profit_product}")
```

```python
# Diference of most and least profitable products

most_profit_product = df["Profit"].max()
least_profit_product = df["Profit"].min()

diff_profit = most_profit_product - least_profit_product

print(f"The profit difference of least and most profitable products is {diff_profit}")
```

```python
# Line plot of costs and profit

fig,ax = plt.subplots(figsize=(12,8))
df.plot(x="Products", y=["Costs","Profit"],
        kind="line", ax=ax)
ax.set(xlabel="Products on Sale",
       ylabel="Amount in US$")
plt.title(label="Cost and Profit Analysis",
          fontdict={"fontsize":20,
                    "fontweight":"bold"})
plt.show()
```

```python
# Average cost per product

avg_cost_product = df["Costs"].mean()
print(f"The average cost per product is {avg_cost_product}")
```

```python
# Average profit per product
avg_profit_product = df["Profit"].mean()
print(f"The average profit per product is {avg_profit_product}")
```
