# Slicing and Analyzing Arrays

```python
# Import necessary libraries

import numpy as np
import matplotlib.pyplot as plt
```

```python
names = ["John", "Kelly", "Jos", "Peter", "Robert", "Piper"]
age = [21,21,56,44,56,96]
gender = ['m', 'f', 'm', 'm', 'm', 'f',]

# Create array of data above
array = np.array([names,age,gender]).T
array
```

```python
# Index last row of array
last_row = array[-1,:]
last_row
```

```python
# Index first two rows of array

first_two_rows = array[:2,:]
first_two_rows
```

```python
# Index age of Peter

age_of_Peter = array[3,1]

print(f'The age of Peter is {age_of_Peter}')
```

```python
# Index of Kelly's age and gender

kelly_age_gender = array[1,1:]
kelly_age_gender
```

```python
# Histogram of ages

# Index ages of array
age_of_array = array[:,1]

# Plot histogram of ages
fig,ax = plt.subplots(figsize=(12,8))
ax.hist(age_of_array,bins=10,rwidth=2.9)
ax.set(xlabel="Ages of People", ylabel="Frequency")
plt.title(label="Age Distribution",fontdict={'fontsize':20, 'fontweight':'bold'})
plt.show()
```
